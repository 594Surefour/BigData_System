```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
path = "./Shanghai_With_R.csv"
dataset = pd.read_csv(path)
del dataset["Unnamed: 0"]
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
#上海市政府/黄浦区
Shanghai_Center_X = 31.2315
Shanghai_Center_Y = 121.1935
dataset['centerDistance']=0.0
for index,row in dataset.iterrows():
    dis = ((row['X']-Shanghai_Center_X)**2) + ((row['Y']-Shanghai_Center_Y)**2)
    #print(dis)
    dataset.loc[index,'centerDistance']=int(dis*1000)
    
```

```python
dataset.head()
```

```python
temp = dataset.Type.sort_values()
```

```python
#Type 的map行为
Type_array = temp.unique()
```

```python
Tpye_map = {}
```

```python
n = 0
for i in Type_array:
    #print(i)
    Tpye_map[n] = i
    n+=1
Tpye_map_new = {value:key for key,value in Tpye_map.items()} 
#Tpye_map_new
#dataset['Type'] = dataset['Type'].map(Tpye_map_new)
```

```python
dataset['Type'] = dataset['Type'].map(Tpye_map_new)
```

```python
dataset.head()
```

```python
dataset.rValue.value_counts()
```

```python
rValue_map = {}
n = 0
for i in range(0,100):
    #print(i)
    rValue_map[n] = (n//10)*10
    n+=1
```

```python
dataset['rValue'] = dataset['rValue'].map(rValue_map)
```

```python
dataset.info()
```

```python
extra = dataset.loc[:,['index','X','Y','District']]
combine = dataset.drop(['index','X','Y','District'],axis = 1)
```

```python
extra.head()
```

```python
combine.head()
```

```python
#combine.to_csv('./train_data.csv')
```

```python
y = combine.loc[:,'rValue']
x = combine.drop(['rValue','rRange'],axis = 1)
```

```python
y
```

```python
x
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
print(x_train.shape,x_test.shape)
print(y_train.shape,y_test.shape)
```

```python
#随机森林
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(x_train, y_train)
#此处产出全部x的结果用于后续打印
y_pred = random_forest.predict(x)
acc_rf = round(random_forest.score(x_train, y_train)*100,2)
acc_rf
```

```python
y_pred
```

```python
predictRValue=pd.DataFrame(y_pred)
```

```python
predictRValue[0].value_counts()
```

```python
predictRValue
```

```python
ans = pd.concat([x,predictRValue],axis=1)
```

```python
ans
```

```python
ans.columns = ['Type', 'centerDistance', 'rValue']
```

```python
ans
```

```python
ans = pd.concat([extra,ans],axis=1)
```

```python
ans['rRange']=0.0
```

```python
ans.head()
```
