```python
pace = 0.001
i=0;
k=0;
for bike_index,bike_row in file.iterrows():
    #print("new bike")
    if i==1000:
        k+=1
        print("mission complete:",k*100000/182834,"%")
        i=0
    startX = round(bike_row['distX'] - pace,6)
    startY = round(bike_row['distY'] - pace,6)
    endX = round(bike_row['distX'] + pace,6)
    endY = round(bike_row['distY'] + pace,6)
    #print(startX,endX)
    #print(startY,endY)
    #弄出单车辐射范围内的地点spot
    spotInBikeCircle = Shanghai.loc[(Shanghai['X']>=startX)&(Shanghai['X']<=endX)&(Shanghai['Y']>=startY)&(Shanghai['Y']<=endY), :]
    if spotInBikeCircle.size != 0:
        #print(spotInBikeCircle)
        
        for spot_index,spot_row in spotInBikeCircle.iterrows():
            
            Shanghai.loc[Shanghai['index']==spot_row['index'],'rValue']+=1
            t_range = round(((bike_row['distX']-spot_row['X'])**2 + (bike_row['distY']-spot_row['Y'])**2)**0.5,4)
            #print(t_range)
            if t_range>spot_row['rRange']:
                Shanghai.loc[Shanghai['index']==spot_row['index'],'rRange'] = t_range
            
        #print("--------------------")
        
    i+=1
```
